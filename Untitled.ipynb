```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('train.csv')
test= pd.read_csv('test.csv')
comb_data = [data,test]
print(data.info())
print(data.shape)
#X=pd.DataFrame(data.groupby(['Pclass'])['Survived'].mean())
pclass=data[['Pclass','Survived']].groupby('Pclass').mean()
#print(X)
#class 1 --> highest survivors
plt.bar([1,2,3],pclass.iloc[:,0])
plt.xticks([1,2,3],('Class1','Class2','Class3'))
```

```python
group_ss=data[['Sex','Survived']].groupby('Sex')
print(group_ss.mean())
for da in comb_data:
    da['Sex']=da['Sex'].replace('female',0)
    da['Sex']=da['Sex'].replace('male',1)
print(test)
```

```python
#data['Age'].fillna(data['Age'].mean(),inplace=True)
for dat in comb_data:
    group_pa=dat[['Pclass','Age']].groupby('Pclass').mean()
    print(type(dat['Age']))
    dat['Age'].fillna(dat['Age'].mean(),inplace=True)
    '''
  m1=dat['Pclass']==1
    m2=dat['Pclass']==2
    m3=dat['Pclass']==3
    dat.loc[m1,'Age']=dat.loc[m1,'Age'].fillna(group_pa.iloc[0][0],inplace=False)
    dat.loc[m2,'Age']=dat.loc[m2,'Age'].fillna(group_pa.iloc[1][0],inplace=False)
    dat.loc[m3,'Age']=dat.loc[m3,'Age'].fillna(group_pa.iloc[2][0],inplace=False)
    #dat[data['Survived']==1]['Age'].hist()
    #dat[data['Survived']==0]['Age'].hist()'''
    
#print(group_pa.iloc[2][0])
data.info()
data.tail()
```

```python
for dat in comb_data:
    dat.loc[ dat['Age'] <= 8, 'Age'] = 0
    dat.loc[(dat['Age'] > 8) & (dat['Age'] <= 16), 'Age'] = 1
    dat.loc[(dat['Age'] > 16) & (dat['Age'] <= 40), 'Age'] = 2
    dat.loc[(dat['Age'] > 40) & (dat['Age'] <= 55), 'Age'] = 3
    dat.loc[(dat['Age'] > 55) & (dat['Age'] <= 80), 'Age'] = 4
    dat.loc[ dat['Age'] > 80, 'Age']
   # dat[['Age', 'Survived']].groupby(['Age'], as_index=False).mean()
```

```python

'''for dat in comb_data:
    dat['AgeBand'] = pd.cut(data['Age'], 5)
#print(data['AgeBand'])
    #dat[['AgeBand', 'Survived']].groupby(['AgeBand'], as_index=False).mean().sort_values(by='AgeBand', ascending=True)
    dat.loc[ data['Age'] <= 16, 'Age'] = 0
    dat.loc[(dat['Age'] > 16) & (dat['Age'] <= 32), 'Age'] = 1
    dat.loc[(dat['Age'] > 32) & (dat['Age'] <= 48), 'Age'] = 2
    dat.loc[(dat['Age'] > 48) & (dat['Age'] <= 64), 'Age'] = 3
    dat.loc[ dat['Age'] > 64, 'Age']=4'''
```

```python
for dat in comb_data:
    dat['family']=dat['SibSp']+dat['Parch']
   # dat[['family', 'Survived']].groupby(['family'], as_index=False).mean().sort_values(by='family', ascending=True)
```

```python
for dat in comb_data:
    #dat[['Embarked', 'Survived']].groupby(['Embarked'], as_index=False).mean().sort_values(by='Embarked', ascending=True)
    dat['Embarked'] = dat['Embarked'].fillna('S')
    dat.loc[dat['Embarked']=='C','Embarked']=3
    dat.loc[dat['Embarked']=='Q','Embarked']=2
    dat.loc[dat['Embarked']=='S','Embarked']=1
```

```python
l=data[['Survived', 'Fare']].groupby(['Survived'], as_index=False).mean()
```

```python
m= l['Fare'].mean()
#print(data['Fare'])
print(m)
for dat in comb_data:
    dat['Fare'].fillna(dat['Fare'].mean(),inplace=True)
    p1=dat['Fare']<=m
    p2=dat['Fare']>m
    dat.loc[p1,'Fare']=0
    dat.loc[p2,'Fare']=1
```

```python
#print(data[data['Fare']==1])
Y_train=data['Survived']
X_train=data.drop(['Name','SibSp','Parch','Ticket','Cabin','Survived','PassengerId'],axis=1)
X_test=test.drop(['Name','SibSp','Parch','Ticket','Cabin','PassengerId'],axis=1)
X_test.info()
```

```python
print(X_test.shape)
print(Y_train.shape)
print(X_train.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import Perceptron
from sklearn.linear_model import SGDClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
logreg = LogisticRegression()
logreg.fit(X_train, Y_train)
Y_pred = logreg.predict(X_test)
acc_log = round(logreg.score(X_train, Y_train) * 100, 2)
acc_log
```

```python
coeff_df = pd.DataFrame(X_train.columns.delete(0))
coeff_df.columns = ['Feature']
coeff_df["Correlation"] = pd.Series(logreg.coef_[0])

coeff_df.sort_values(by='Correlation', ascending=False)
```

```python
# Support Vector Machines
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import Perceptron
from sklearn.linear_model import SGDClassifier
from sklearn.tree import DecisionTreeClassifier
svc = SVC()
svc.fit(X_train, Y_train)
Y_pred = svc.predict(X_test)
acc_svc = round(svc.score(X_train, Y_train) * 100, 2)
acc_svc



knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train, Y_train)
Y_pred = knn.predict(X_test)
acc_knn = round(knn.score(X_train, Y_train) * 100, 2)
acc_knn


gaussian = GaussianNB()
gaussian.fit(X_train, Y_train)
Y_pred = gaussian.predict(X_test)
acc_gaussian = round(gaussian.score(X_train, Y_train) * 100, 2)
acc_gaussian

perceptron = Perceptron()
perceptron.fit(X_train, Y_train)
Y_pred = perceptron.predict(X_test)
acc_perceptron = round(perceptron.score(X_train, Y_train) * 100, 2)
acc_perceptron

linear_svc = LinearSVC()
linear_svc.fit(X_train, Y_train)
Y_pred = linear_svc.predict(X_test)
acc_linear_svc = round(linear_svc.score(X_train, Y_train) * 100, 2)
acc_linear_svc


decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_train, Y_train)
Y_pred = decision_tree.predict(X_test)
acc_decision_tree = round(decision_tree.score(X_train, Y_train) * 100, 2)
acc_decision_tree

random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(X_train, Y_train)
Y_pred = random_forest.predict(X_test)
random_forest.score(X_train, Y_train)
acc_random_forest = round(random_forest.score(X_train, Y_train) * 100, 2)
acc_random_forest

models = pd.DataFrame({
    'Model': ['Support Vector Machines', 'KNN', 'Logistic Regression', 
              'Random Forest', 'Naive Bayes', 'Perceptron', 'Linear SVC', 
              'Decision Tree'],
    'Score': [acc_svc, acc_knn, acc_log, 
              acc_random_forest, acc_gaussian, acc_perceptron, acc_linear_svc, acc_decision_tree]})
models.sort_values(by='Score', ascending=False)
```
